```python
a,b,c=map(float,input().split())
if a>0 and b>0 and c>0 and a+b>c and a+c>b and b+c>a :
    A=(a+b+c)/2
    B=A*(A-a)*(A-b)*(A-c)
    C=B**(1/2)
    print('%.1f'%C)
else:
    print('khong phai 3 canh tam giac')
```
